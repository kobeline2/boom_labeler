```python
import cv2
import math
import matplotlib.pyplot as plt
import numpy as np
```

### [read image] 

```python
# time lapse camera image
im_src = cv2.imread("./ref_img/img_src.tiff")
h, w, c = im_src.shape
print(h, w, c)
plt.imshow(im_src)
plt.show()
```

```python
# reference image
im_dst = cv2.imread("./ref_img/img_dst.tiff")
h, w, c = im_dst.shape
print(h, w, c)
plt.imshow(im_dst)
plt.show()
```

### [AffineTransform]

```python
src_pts = np.array([[831, 425], [166, 511], [1070, 430]], dtype=np.float32)
dst_pts = np.array([[426, 334], [710, 581], [553, 250]], dtype=np.float32)

mat = cv2.getAffineTransform(src_pts, dst_pts)
affine_img = cv2.warpAffine(im_src, mat, (w, h))
plt.imshow(affine_img)
plt.show()
```

### [PerspectiveTransform]

```python
src_pts = np.array([[831, 425], [166, 511], [1070, 430], [164, 508]], dtype=np.float32)
dst_pts = np.array([[426, 334], [710, 581], [553, 250], [164, 508]], dtype=np.float32)
mat = cv2.getPerspectiveTransform(src_pts, dst_pts)
perspective_img = cv2.warpPerspective(im_src, mat, (w*4, h*4))
# cv2.imwrite('data/dst/opencv_perspective_dst.jpg', perspective_img)
plt.imshow(perspective_img)
plt.show()
```
